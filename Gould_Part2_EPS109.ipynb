#### Now plot the deviation of every single data set from the solution into one diagram. Make some polite comments about the most negative and the most positive deviations that you see.

```python
differences = []

nFiles = 0;

for root, dirs, files in os.walk(folder):  
    for filename in files:
        if (filename.find('.txt')>=0):
            xi = np.loadtxt(folder+'/'+filename)
            difference = xi - solutions
            plt.plot(difference)
            differences.append(difference)
            nFiles += 1;

differences = np.array(differences)
print(np.shape(differences) )    

print('Most positive deviation: ', np.max((differences)))
print('Most negative deviation: ', np.min((differences)))
plt.show()
```

The most negative difference is .96 away from the solution, the most positive difference is .94 away from the solution. Since the scale is only out of 1, being .94 away is very far, indicating someone may have skipped that question and guessed, or may have imputted it in wrong.

#### Compute the variance between every data set and the solution. Store all variances in a Numpy vector 'var'.

```python
nFiles = 0;

var = []
for root, dirs, files in os.walk(folder):  
    for filename in files:
        if (filename.find('.txt')>=0):
            xi = np.loadtxt(folder+'/'+filename)
            nNumbers = len(xi)
            nFiles += 1;
            difference = xi - solutions
            average_of_difference = np.sum(difference)/nNumbers
            variance = np.sqrt(average_of_difference**2 / (nNumbers -1))
            var.append(variance)
var = np.array(var)
print(var)
```

#### Plot the variances stored in 'var', determine the file name of data sets with the largest and smallest variance.

```python
plt.plot(var)
plt.title('Variance Between Every Data Set and the Solution')

for root, dirs, files in os.walk(folder):
    print('Largest Variance: ', files[np.argmax(var)], np.max(var))
    print('Smallest Variance: ', files[np.argmin(var)], np.min(var))


plt.show()
```

#### What is the rank of your data set in an ordered list of all variances?

```python
my_variance = 0.0004221158824088691

ordered_var = np.sort(var)

for i in range(len(ordered_var)):
    if ordered_var[i] == my_variance:
        print('My rank = ', i,'th best guesser in the class')
```

#### Compute and print the variance, sigma, in the variable among all deviations stored in the variable 'diff'

```python
average_of_diff = np.sum(diff)/len(diff)
print('Mean: ', average_of_diff)
```

```python
variance = np.sqrt(average_of_diff**2 / (len(diff) -1))
print('Variance: ', variance)
```

```python
sigma = variance / np.sqrt(len(diff))
print('Sigma: ', sigma)
```

#### Replot the probability density from above and overplot a Gaussian Bell curve with the variance that you just computed.

```python
sd = np.sqrt(variance)

x = np.linspace(-.5, .5, len(diff))

dist = 1.0/(sd*np.sqrt(2*np.pi))*np.exp(-(x - average_of_diff)**2/(2*sd**2)) 

plt.plot(x, dist*20)

plt.hist(diff, bins='auto')
plt.xlim(-.5, .5)
plt.title('Probablility Distribution of Deviations')
plt.show()
```
