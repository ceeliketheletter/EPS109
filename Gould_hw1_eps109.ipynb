## (7) Plotting the world’s CO2 record (6pts):

In the lab, we practiced plotting the world’s temperature data. Now we want to repeat these
steps with the world’s CO2 data that is stored in columns 1 and 2 in a file
ice_core_CO2_data.txt. Download this file and generate a plot CO2 contents versus
time. As in the lab, add a shaded region that illustrates the deviations from the averaged
CO2 contents

```python
data = np.loadtxt('ice_core_CO2_data(1).txt', usecols=(0,1))

plt.rcParams['figure.figsize'] = [15, 5]
plt.plot(data[:,0],data[:,1],'r-',linewidth=1,markersize=0,mec='r',mew=2, mfc='pink')
plt.xlabel('Time')
plt.ylabel('CO2')
plt.show()
```

```python
CO2 = data[:,1]
print(sum(CO2))
print(CO2.shape)
average = sum(CO2) / CO2.shape[0]
print(average)
```

```python
plt.rcParams['figure.figsize'] = [15, 5]
plt.plot(data[:,0],data[:,1],'r-',linewidth=1,markersize=0,mec='r',mew=2, mfc='pink');
plt.fill_between(data[:,0],data[:,1],average,color=(1.0,0.90,0.90),lw=0)
plt.xlabel('Time')
plt.ylabel('CO2')
plt.show()
```
