The Mandelbrot and Julia sets both rely on the very same complex iteration formula: z(n+1) = (zn)^2 + C. For Mandelbrot set, you always start with z0=0 and C is the parameter that you vary. The Mandelbrot set is defined as set of all points C in the complex plane for which the series (#) does
not diverge

Conversely for the Julia set, the roles of z0 and C are reversed. For this set, you vary the starting number, z0=x+y×i. The parameter C stays fixed at some complex number to be specified beforehand. For every C value, a somewhat different Julia set emerges. The Julia set is defined as
the set of complex points z0, for which the series (#) does not diverge.

```python
import numpy as np
import matplotlib.pyplot as plt

nmax = 50
zlimit = 50.0
N = 201

series = []                    #create an empty list for answers

x_range = np.linspace(-1.5, 1.5, N)  #xrange we are parsing over
y_range = np.linspace(-1.5, 1.5, N) # yrange we are parsing over

c = complex(-0.4, 0.6)          #c is predetermined

def julia(x,y):                  #define a function that takes x and y
    z = complex(x,y)            #we have a starting number z0 that is complex, calling x and y
    
    for n in range(0,nmax):    #for every value from 0-50
        z = z**2 + c            #new z is z0 squared plus c
        if abs(z) >= zlimit:    #if new z is over the limit, it diverges
            return n            #returm the value at which it diverged
        
        

for x in x_range:            #for every value in x range
    for y in y_range:         #for every value in y range
        
         series.append(julia(x,y)) #apply the function for this given x and y and append to our answers list
        
         if julia(x,y) == None:    #if the value does not diverge, it converges and is represented by "None"
            series.remove(None)     #remove the "None"s
            series.append(nmax)      #replace with nmax=50
        
series = np.array(series)      #convert our list of answers to an array
series.shape = (N,N)           #reshape our array into matrix format
```

```python
plt.imshow(series, extent=(-1.5, 1.5, -1.5, 1.5))
plt.title("Julia Set for C = %s" %c )
#plt.savefig('gould_julia_1.png', dpi = 300, bbox_inches='tight')
plt.show()
```
