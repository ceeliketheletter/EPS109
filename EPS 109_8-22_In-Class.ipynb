# (6) Let us generate a 3D plot

```python
n=51
L = 10.0
x = np.arange(-L, L, L/n)
y = np.arange(-L, L, L/n)
X, Y = np.meshgrid(x, y)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
R = np.sqrt(X**2+Y**2) + 1e-14
Z = np.sin(R) / R

plt.plot(Z)
plt.show()
```

```python
#this requires some extra libraries
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')  #this creates the 3d box
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False) #this is the plot inside it
plt.show()
```
