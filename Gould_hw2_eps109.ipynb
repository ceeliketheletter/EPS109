```python
import numpy as np
import matplotlib.pyplot as plt

n = 500
r = n/2.0
r = np.int(r)

data = np.zeros((n,n))



for i in range(0, n):
   for j in range(0,n):
    #print(i,j) 
        if i < np.int(n/2):
            i_center = 0
        if i > np.int(n/2):
            i_center = n
        if j < np.int(n/2):
            j_center = 0
        if j > np.int(n/2):
            j_center = n
        distance = (i - i_center)**2 + (j-j_center)**2
        if distance < r*r:
            data[i,j] = 1





plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 500
r = n/3.0
r = np.int(r)

data = np.zeros((n,n))



for i in range(0, n):
   for j in range(0,n):
    #print(i,j) 
        i_center = np.int(n/2)
        
        j_center_1 = r
        
        j_center_2 = 2*r
        
        distance_1 = (i - i_center)**2 + (j-j_center_1)**2
        distance_2 = (i - i_center)**2 + (j-j_center_2)**2
        
        if distance_1 < r*r:
            data[i,j] = 1
        
        if distance_2 < r*r:
            data[i,j] = 1
        
        if distance_1 < r*r and distance_2 < r*r:
            data[i,j] = 0
            
    

plt.imshow(data, interpolation='nearest')
plt.show()
```
