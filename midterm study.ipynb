```python
n = 500
x = np.zeros((n,n))
r = n//2

for i in range(n):
    for j in range(n):
        area = ((i-r)**2+ (j-r)**2) * np.pi
        if area < r**2 * np.pi:
            x[i,j] = 1
plt.imshow(x)
```
