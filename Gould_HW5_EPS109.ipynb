Now define your own function LM(x,r)that returns the logistic map function f(x) = r × x × (1-x). Define a second function LM2(x,r)that returns f2(x)=f(f(x)). Define a third and a forth function, LM3(x,r) andLM4(x,r)that returns f3(x)=f(f(f(x)))and f4(x)=f(f(f(f(x)))). Together with the linear function y=x, plot f(x), f2(x), f3(x)and f4(x) as function of x in one diagram

```python
def LM(x,r):
    fx = r * x * (1-x)
    return fx

def LM2(x,r):
    f2x = LM(LM(x,r),r)
    return f2x

def LM3(x,r):
    f3x = LM(LM(LM(x,r),r),r)
    return f3x

def LM4(x,r):
    f4x = LM(LM(LM(LM(x,r),r),r),r)
    return f4x
```
