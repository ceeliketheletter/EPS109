```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Newtons_Equation(X):
    r = X[0:2]
    v = X[2:4]

    
    F = -G * ms * mp /np.linalg.norm(r)**3 * r
    a    = F / mp
    
    return np.concatenate((v, a))
```

```python
h = .001
time = np.arange(0,10.001, h)

ms = 3e+6
mp = 1
G  = 1e-6
```

```python
x = 1
y = 0
vx = 0
vy = 2

r = np.array([x,y])
v = np.array([vx,vy])


xt = []
yt = []
vx = []
vy = []
tt = []

t = 0


X = np.concatenate((r,v))

while(t < 10):

    r = X[0:2]
    v = X[2:4]
    
    
    xt.append(X[0])
    yt.append(X[1])
    
    X = X + Newtons_Equation(X) * h
    
    tt.append(t)
    t = t + h
    
    
    
plt.plot(xt, yt, 'r-')
plt.plot(0,0,'*',mfc='w',ms=10)
#plt.gca().set_aspect('equal', adjustable='box')
plt.show()

plt.plot(tt,xt, 'r-')
plt.show()
```
