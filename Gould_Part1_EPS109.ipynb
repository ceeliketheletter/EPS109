```python
#Load the Numpy and MatPlotLib packages as usual but also 'os' so that we cazn read many files
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
solutions = np.loadtxt('x_positions_used_in_lecture.txt')
```

```python
my_answer = np.loadtxt('data_entry_sheet.txt')
```

```python
nNumbers = len(my_answer)
```

```python
plt.plot(my_answer, label='my answer', marker='o')
plt.plot(solutions, label = 'solutions')
plt.title('My Answer and Solutions')
plt.legend()
plt.show()
```

```python
difference = my_answer - solutions
plt.plot(difference)
line = np.zeros(nNumbers)
plt.plot(line)
plt.title('Difference between My Answer and Solutions')
plt.show()

print('Most Positive Deviation: ', np.max(difference))
print('Most Negative Deviation: ', np.min(difference))
```

This highest deviation on the positive side is larger than the highest deviation on the negative side, possibly because I was using 0 and .5 and a reference, so the lower numbers will be more accurate.

```python
average_of_difference = np.sum(difference)/nNumbers
print(average_of_difference)
```

The sign of the average is positive because the data in 'my_answers' are larger than 'solutions'

```python
variance = np.sqrt(average_of_difference**2 / (nNumbers -1))
print('Variance: ', variance)
sigma = variance / np.sqrt(nNumbers)
print('Sigma: ', sigma)
```
